```python
#Version using Jeff's contours for the colors 

plt.figure(figsize=(14,11))

#Temperature and updrafts--weak storms
plt.subplot(2,2,1)
plt.contourf(mat['R'][0,:], mat['lev'][:,0], mat['Temp_18_21'], 
             levels=np.linspace(-11,11,23), cmap=plt.cm.RdBu_r)
plt.gca().invert_yaxis()
plt.gca().set_xlim([0,400])
plt.gca().set_ylim([1000,80])
cbar = plt.colorbar()
cbar.ax.set_ylabel('Warm core anomaly (K)', fontsize=12)
CS = plt.contour(mat['R'][0,:], mat['lev'][:,0], mat['Omega_18_21']*100,
                 levels=np.linspace(-2.1,2.1,15), colors='k')
plt.xlabel('Radius (km)', fontsize=12)
plt.ylabel('Pressure (hPa)', fontsize=12)
plt.title('(a) Temperature anomaly and pressure velocity :\n18-21 m/s intensity', fontsize=14)


#Velocities--weak storms
plt.subplot(2,2,2)
plt.contourf(mat['R'][0,:], mat['lev'][:,0],mat['Tan_V_18_21'], 
             levels=np.linspace(-50,50,21), cmap=plt.cm.RdBu_r)
plt.gca().invert_yaxis()
plt.gca().set_xlim([0,1000])
plt.gca().set_ylim([1000,80])
cbar = plt.colorbar()
cbar.ax.set_ylabel('Tangential velocity (m/s)', fontsize=12)
CS = plt.contour(mat['R'][0,:], mat['lev'][:,0],mat['Rad_V_18_21'],
                 levels=np.linspace(-21,21,15), colors='k')
plt.xlabel('Radius (km)', fontsize=12)
plt.ylabel('Pressure (hPa)', fontsize=12)
plt.title('(b) Tangential and radial velocities:\n18-21 m/s intensity', fontsize=14)


#Temperature and updrafts--strong storms
plt.subplot(2,2,3)
plt.contourf(mat['R'][0,:], mat['lev'][:,0], mat['Temp_30_33'], 
             levels=np.linspace(-12,12,25), cmap=plt.cm.RdBu_r)
plt.gca().invert_yaxis()
plt.gca().set_xlim([0,400])
plt.gca().set_ylim([1000,80])
cbar = plt.colorbar()
cbar.ax.set_ylabel('Warm core anomaly (K)', fontsize=12)
CS = plt.contour(mat['R'][0,:], mat['lev'][:,0], mat['Omega_30_33']*100,
                 levels=np.linspace(-2.1,2.1,15), colors='k')
plt.xlabel('Radius (km)', fontsize=12)
plt.ylabel('Pressure (hPa)', fontsize=12)
plt.title('(c) Temperature anomaly and pressure velocity :\n30-33 m/s intensity', fontsize=14)


#Velocities--strong storms
plt.subplot(2,2,4)
plt.contourf(mat['R'][0,:], mat['lev'][:,0],mat['Tan_V_30_33'], 
             levels=np.linspace(-50,50,21), cmap=plt.cm.RdBu_r)
#plt.gca().clabel(CS, CS.levels, inline=False)
plt.gca().invert_yaxis()
plt.gca().set_xlim([0,1000])
plt.gca().set_ylim([1000,80])
cbar = plt.colorbar()
cbar.ax.set_ylabel('Tangential velocity (m/s)', fontsize=12)
CS = plt.contour(mat['R'][0,:], mat['lev'][:,0],mat['Rad_V_30_33'],
                 levels=np.linspace(-21, 21, 15), colors='k')
plt.xlabel('Radius (km)', fontsize=12)
plt.ylabel('Pressure (hPa)', fontsize=12)
plt.title('(d) Tangential and radial velocities:\n30-33 m/s intensity', fontsize=14)


plt.tight_layout()
plt.savefig('plots/radial_vertical_profiles_V1_Jeff_contours.png')

#Multiplied pressure velocity by 100 (Jeff did this too)--
#must have been in hPa per second, this gets it in Pa/s and comparable to the other models in Moon et al.
```
